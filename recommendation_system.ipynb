```python
movies = movies.merge(credits,on='title')
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.dropna(inplace=True)
```

```python
import ast
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.iloc[0]['genres']
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.iloc[0]['keywords']
```

```python
movies.head()
```

```python
def convert3(obj):
    L = []
    count = 0
    for i in ast.literal_eval(obj):
        if count != 3:
            L.append(i['name'])
        count+=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head(5)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(5)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ", ""))
    return L1
```

```python
movies['overview'] = movies['overview'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
```

```python
movies.head(5)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head(5)
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new.head(5)
```

```python
new['tags'] = new['tags'].apply(lambda x:" ".join(x))
```

```python
new.head(5)
```

```python
new['tags'] = new['tags'].apply(lambda x:x.lower())
```

```python
new
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(word):
    L = []
    for i in word.split():
        L.append(ps.stem(i))
    return " ".join(L)
```

```python
new['tags'] = new['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000 , stop_words='english')
```

```python
vectors = cv.fit_transform(new['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
simalirity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(simalirity[0])), reverse= True , key= lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new[new['title'] == movie].index[0]
    distance = simalirity[movie_index]
    movie_list = sorted(list(enumerate(distance)), reverse= True , key= lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(new.iloc[i[0]].title)
```

```python
recommend('The Lone Ranger')
```
